# Exercise 5: Using NumPy and Matplotlib libraries

```python
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
# Titles/Labels:
plt.title("Exercise 5")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.ylim(0, 1)
plt.xlim(0, 1)

# purple table (using one array seperated into two by transpose)
purple_data =[(0, 0), (0, 0), (0, 0), (0.25, 0.25), (0.5, 0.4), (0.5, 0.33), 
              (0.5, 0.29), (0.75,0.38), (0.75, 0.33), (1, 0.4)]
purple_data = np.asarray(purple_data)
plt.plot(*purple_data.transpose(), color="Purple", label="Purple Chart/Blue Line")
plt.plot(*purple_data.transpose(), 'r*', color="Purple")

# blue table (using two arrays, one for x one for y)
blue_data_x = [0.25, 0.25, 0.25, 0.5, 0.5, 0.75, 0.75, 0.75, 1, 1]
blue_data_y = [1, 0.5, 0.33, 0.5, 0.4, 0.5, 0.43, 0.38, 0.44, 0.4]
plt.plot(blue_data_x, blue_data_y, color="Blue", label="Blue Chart/Purple Line")
plt.plot(blue_data_x, blue_data_y, 'r*', color="Blue")

plt.legend()
plt.show()
```
